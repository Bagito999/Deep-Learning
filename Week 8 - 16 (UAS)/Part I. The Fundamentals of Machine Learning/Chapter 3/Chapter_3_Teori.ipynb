#  Chapter 3 - Classification

##  Pengantar

Dalam Machine Learning, klasifikasi adalah tugas untuk memprediksi label dari sebuah instance. Berbeda dengan regresi (yang memprediksi nilai kontinu), klasifikasi menghasilkan kelas diskret. Contoh klasik adalah memprediksi angka dari gambar tangan pada dataset **MNIST**.

---

##  Dataset MNIST

MNIST adalah kumpulan gambar digit tangan (0-9) sebanyak 70.000 gambar resolusi 28×28 piksel (784 fitur). Setiap gambar dilabeli dengan digit yang diwakilinya.

- 60.000 data untuk training
- 10.000 data untuk testing

Fitur berupa intensitas piksel (0 = putih, 255 = hitam).

---

##  Training Binary Classifier

Awalnya, masalah disederhanakan menjadi klasifikasi **biner**, misalnya: mendeteksi angka **5** (5 vs bukan 5).

Langkah-langkah:
- Label dikonversi menjadi boolean: `True` jika 5, `False` jika bukan.
- Model seperti **SGDClassifier** digunakan karena efisien untuk dataset besar.

---

##  Pengukuran Kinerja

### 1. **Akurasi**

Proporsi prediksi yang benar.

\[
$\text{Akurasi} = \frac{\text{Jumlah prediksi benar}}{\text{Total prediksi}}$
\]

 Tidak cocok untuk data imbalanced.

---

### 2. **Confusion Matrix (Matriks Kebingungan)**

Menampilkan frekuensi prediksi benar dan salah dari setiap kelas.

|              | Prediksi Positif | Prediksi Negatif |
|--------------|------------------|------------------|
| **Positif**  | True Positive     | False Negative   |
| **Negatif**  | False Positive    | True Negative    |

---

### 3. **Precision dan Recall**

- **Precision**: Ketepatan prediksi positif

\[
$\text{Precision} = \frac{\text{TP}}{\text{TP + FP}}$
\]

- **Recall**: Kemampuan mendeteksi semua positif

\[
$\text{Recall} = \frac{\text{TP}}{\text{TP + FN}}$
\]

- **F1 Score**: Harmonik dari Precision dan Recall

\[
$\text{F1} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$
\]

---

### 4. **Trade-off Precision vs Recall**

Ada kompromi: menaikkan threshold prediksi bisa meningkatkan precision tapi menurunkan recall, dan sebaliknya.

---

### 5. **ROC Curve (Receiver Operating Characteristic)**

- Menampilkan **True Positive Rate (TPR)** vs **False Positive Rate (FPR)**.
- Area di bawah kurva (AUC) = metrik kinerja.

\[
$\text{TPR} = \text{Recall} = \frac{\text{TP}}{\text{TP + FN}}$
\]
\[
$\text{FPR} = \frac{\text{FP}}{\text{FP + TN}}$
\]

---

##  Multiclass Classification

Klasifikasi lebih dari dua kelas (0–9). Dua strategi utama:

- **One-vs-Rest (OvR)**: Latih satu classifier per kelas (10 total). Pilih kelas dengan skor tertinggi.
- **One-vs-One (OvO)**: Latih classifier untuk setiap pasang kelas (45 total). Kelas dengan kemenangan terbanyak dipilih.

Scikit-Learn menangani ini secara otomatis.

---

##  Analisis Error

Langkah-langkah:

1. Gunakan **confusion matrix** untuk analisis kesalahan.
2. Plot matrix sebagai gambar untuk identifikasi pola.
3. Fokus pada kesalahan signifikan (misalnya 3 sering diprediksi sebagai 5).
4. Normalisasi confusion matrix untuk melihat proporsi kesalahan.

---

##  Multilabel Classification

Satu instance bisa memiliki **lebih dari satu label**.

Contoh:
- Deteksi apakah angka ≥ 7 (**label 1**)
- Deteksi apakah angka ganjil (**label 2**)

Evaluasi dengan **macro-average F1 score**:

\[
$\text{F1}_{macro} = \frac{1}{L} \sum_{i=1}^{L} \text{F1}_i$
\]

(L = jumlah label)

---

##  Multioutput Classification

Jenis klasifikasi multilabel, tetapi **setiap label dapat memiliki nilai lebih dari dua kelas**.

Contoh:
- Input: gambar angka dengan noise
- Output: gambar bersih
- Setiap piksel diperlakukan sebagai label multi-kelas (nilai 0–255)

Ini bisa dianggap sebagai klasifikasi atau regresi tergantung konteksnya.

---

##  Kesimpulan

- Klasifikasi membutuhkan evaluasi lebih kompleks dibanding regresi.
- Banyak metrik penting: precision, recall, F1, ROC, AUC.
- Multi-class, multilabel, dan multioutput perlu pendekatan berbeda.
- Analisis error dan preprocessing data (seperti alignment gambar) penting untuk peningkatan performa.
