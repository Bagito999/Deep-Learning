#  Chapter 1 - The Machine Learning Landscape

##  Apa Itu Machine Learning?

**Definisi umum:**
> Machine Learning adalah ilmu (dan seni) dalam memprogram komputer agar dapat belajar dari data.

**Definisi oleh Arthur Samuel (1959):**
> "Field of study that gives computers the ability to learn without being explicitly programmed."

**Definisi oleh Tom Mitchell (1997):**
> "A computer program is said to learn from experience **E** with respect to some task **T** and some performance measure **P**, if its performance on **T**, as measured by **P**, improves with experience **E**."

**Contoh:**
- Sistem filter spam:
  - **T**: Mengklasifikasi email sebagai spam atau bukan
  - **E**: Dataset email spam dan ham
  - **P**: Akurasi klasifikasi email

##  Mengapa Menggunakan Machine Learning?

Machine Learning sangat berguna untuk:
- Masalah yang terlalu kompleks untuk diprogram manual
- Lingkungan data yang berubah-ubah (adaptif)
- Menemukan pola tersembunyi (data mining)

---

##  Jenis-Jenis Sistem Machine Learning

### 1. Supervised Learning
Model belajar dari **data berlabel** (label sudah diketahui).
- Contoh:
  - Klasifikasi (Spam/Not Spam)
  - Regresi (Harga rumah)

**Algoritma populer:**
- Linear Regression
- Logistic Regression
- Decision Tree
- Random Forest
- Neural Network

### 2. Unsupervised Learning
Model belajar dari **data tak berlabel**.
- Contoh:
  - Clustering (pengelompokan)
  - Deteksi anomali
  - Visualisasi data

**Algoritma populer:**
- K-Means
- DBSCAN
- PCA, t-SNE
- Apriori, Eclat (association rules)

### 3. Semisupervised Learning
Model belajar dari campuran data berlabel dan tak berlabel.
- Contoh: Google Photos â€“ mengenali wajah dengan 1-2 label saja

### 4. Reinforcement Learning
Agen belajar dari interaksi dengan lingkungan dan menerima **reward** atau **penalty**.
- Contoh: AlphaGo, robot berjalan

##  Tantangan dalam Machine Learning

1. Data training tidak cukup
2. Data tidak representatif
3. Data berkualitas buruk (noise, outlier)
4. Fitur tidak relevan
5. **Overfitting** â€“ terlalu pas dengan data training
6. **Underfitting** â€“ model terlalu sederhana

**Solusi Overfitting:**
- Tambah data
- Regularisasi
- Model yang lebih sederhana

##  Evaluasi dan Validasi

- **Training Set**: untuk melatih model
- **Validation Set**: untuk memilih dan tuning model
- **Test Set**: untuk evaluasi akhir

>  Hindari kebocoran data antara training dan test set.

---

##  Hyperparameter vs Parameter

- **Parameter**: dipelajari dari data (misalnya bobot dalam regresi)
- **Hyperparameter**: ditentukan sebelum pelatihan (misal: learning rate, jumlah neighbor `k`)
